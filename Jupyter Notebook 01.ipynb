### This code will calculate the TDEE (total daily energy expenditure) using gender, age, weight, height and activity level and the required daily calories to lose and bulk using the Mifflin-St Jeor formula.

```python
gender = input('Gender (M/F) : ')
age = float(input('Age : '))
weight = float(input('Weight (in kg) : '))
height = float(input('Height (in cm) : '))
level= float(input('Activity Level \n1 - Sedentary. \n2 - Light Exercise (1-2 times/week) \n3 - Moderate (3-5 times/week) \n4 - Heavy (6-7 times/week) \n5 - Athlete (2 times/day) :'))

if gender == 'M':
    BMR = (10 * weight) + (6.25 * height) - (5 * age) + 5
if gender == 'F':
    BMR = (10 * weight) + (6.25 * height) - (5 * age) - 161
    



if level == 1:
    activity = 1.2
if level == 2:
    activity = 1.375
if level == 3:
    activity = 1.55
if level == 4:
    activity = 1.725
if level == 5:
    activity = 1.9

TDEE = float((BMR * activity))
print('\nThe maintenance calories is', TDEE)
print('For cutting calories, the daily calorie intake is', float(TDEE - 500))
print('For bulking calories, the daily calorie intake is', float(TDEE + 500))
```
